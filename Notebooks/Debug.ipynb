```python
import sys

def print_multiline_progress(lines, clear_previous=True):
    """
    Print multiple lines that can be updated in place.
    
    Parameters:
        lines (list of str): Lines to print.
        clear_previous (bool): If True, moves the cursor up to overwrite the previous block.
                                Set this to False on the first call.
    """
    if clear_previous:
        # Move the cursor up by the number of lines that were printed previously.
        sys.stdout.write("\033[F" * len(lines))
    # Print the updated lines.
    sys.stdout.write("\n".join(lines) + "\n")
    sys.stdout.flush()

# Example usage:
import time

# Let's assume the progress block has 3 lines.
progress_lines = [
    "Progress: 0%",
    "Time elapsed: 0s",
    "Info: Starting..."
]

# First call: print without moving up.
print_multiline_progress(progress_lines, clear_previous=False)

for i in range(1, 11):
    # simulate new progress information
    progress_lines = [
        f"Progress: {i*10}%",
        f"Time elapsed: {i}s",
        f"Info: Step {i}"
    ]
    time.sleep(0.5)
    print_multiline_progress(progress_lines)  # clear_previous defaults to True
```

```python
from IPython.display import clear_output
import time

def print_multiline_progress_jupyter(lines):
    clear_output(wait=True)  # Clear the current cell output.
    print("\n".join(lines))

# Example usage:
for i in range(10):
    progress_lines = [
        f"Progress: {i*10}%",
        f"Time elapsed: {i}s",
        f"Info: Step {i}"
    ]
    print_multiline_progress_jupyter(progress_lines)
    time.sleep(0.5)
```

```python
from IPython.display import display, update_display
import time

progress_text = "Starting..."
display_handle = display(progress_text, display_id=True)

for i in range(10):
    progress_text = f"Progress: {i*10}%\nTime elapsed: {i}s\nInfo: Step {i}"
    update_display(progress_text, display_id=display_handle.display_id)
    time.sleep(0.5)
```
