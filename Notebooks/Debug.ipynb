```python
from importing import *
model = load_model("higgs")

# Values
v = 246

# Tadpole conditions
Z1, R_Z6, I_Z6 = 1, 1, 1
Y11 = -v**2/2*Z1
R_Y12 = v**2/2*R_Z6
I_Y12 = v**2/2*I_Z6
# Mass parameter reassign
MHp, Z3 = 500, 4
Y22 = MHp**2 - (v**2/2)*Z3

V0_params_values = [Y11, Y22,
                    Z1,1,Z3,4,
                    R_Y12,I_Y12,
                    -1,-1,R_Z6,I_Z6,-1,1]

subs_V0_params_values = {symb:val for symb, val in zip(model.V0_params, V0_params_values)}

subs_VEV_values = {model.symbol("v"): v}


print(subs_VEV_values, subs_V0_params_values)

# ModelDataCalculator
calc = ModelDataCalculator(model, subs_V0_params_values, subs_VEV_values)
```
